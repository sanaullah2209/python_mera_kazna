```python
class Student():
    #initailaiser / constrcutor
    # ya hamre pas specail methoad ha 
    # ya value ap ko dene zarraor ha ap kese ko be miss nahi kar sakte 
    def __init__(self,name,age,city,course="python"):
        self.name = name
        self.age = age 
        self.city = city
        self.course = course
   
    # pehle hum ise function kahte te magar agar class ki andar hoga to use hum methods kahe gi 
    # function / methods
    def apper_in_exam(self):
        print(f"{self.name} is appearing in exam ")
   
    def pay_fees(self):
        print(f"{self.name} is paying the fees  ")
```

```python
s4 = Student()
```

```python
s4 = Student('Taha',20 ,'Karachi', 'Python')
```

```python
s5 = Student('imrankhan',70 ,'islambad', 'datascience')
```

```python
s4.apper_in_exam()
```

```python
#what is self 
#self kud object hota ha 
# self ma kud object chala jata ha
#self copy banta ha who object ki copy banta ha 
#self ki jaga hum kuch or be name de sakte ha 

class Car():
    def __init__(self,model,make,color="black "): # ya default value ha agar koi value nahi mile gi to ya default de dega
        self.model = model
        self.make = make
        self.color = color
        # hard cotet value dal sakte ha ap 
        self.ac = "Dawlance"
    def description_car(self):
        print(f"""
        Model Name : {self.model}
        Car Make : {self.make}
        Car color :{self.color}
        Car ac :{self.ac}
        """)
```

```python
c1 = Car('2024',"Honda",'black')
c2 = Car('2025',"Honda",'White')
```

```python
c1.description_car()
```

```python
c2.description_car()
```

```python
#call hemehsa is tara karna chaye

Car.description_car(c1)
```
